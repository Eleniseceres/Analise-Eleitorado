# Analise do Eleitorado brasileiro


```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carregar o arquivo csv
eleitorado = pd.read_csv('C:/Users/eleni/Documents/eleitorado_municipio_2018.csv', sep=';', encoding='latin1')
```

```python
#5 cidades com maior quantidade de deficientes
eleitorado.nlargest(5, 'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Eleitoras', eleitorado['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores', eleitorado['QTD_ELEITORES_MASCULINO'].sum())
print('Não Informado', eleitorado['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
#Variáveis para calcular percentual
totaleleitores = eleitorado['QTD_ELEITORES'].sum()
totalfeminino = eleitorado['QTD_ELEITORES_FEMININO'].sum()
totalmasculino = eleitorado['QTD_ELEITORES_MASCULINO'].sum()
totalnaoinformado = eleitorado['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
#Mostrar valores percentual
#round arredonda as casas decimais
print('Eleitoras:', ((totalfeminino / totaleleitores) * 100).round(2))
print('Eleitores:', ((totalmasculino / totaleleitores) * 100).round(2))
print('Não Informado:', ((totalnaoinformado / totaleleitores) * 100).round(2))
```

```python
#Municipios com mais homens que mulheres
eleitorado[eleitorado['QTD_ELEITORES_MASCULINO'] > eleitorado['QTD_ELEITORES_FEMININO']].count()
```

```python
#Criar coluna indicando relação feminino e masculino
eleitorado['RELACAO_FE_MASC'] = eleitorado['QTD_ELEITORES_FEMININO'] / eleitorado['QTD_ELEITORES_MASCULINO']
```

```python
eleitorado.nlargest(5, 'RELACAO_FE_MASC').T
```

```python
#Menor relação feminino e masculino
eleitorado.nsmallest(5, 'RELACAO_FE_MASC').T
```

```python
#Retirando eleitores que estão no exterior
eleitoradobrasil = eleitorado[eleitorado['CD_PAIS']==1].copy()
```

```python
eleitoradobrasil.head().T
```

```python
#Apenas eleitores que estão no Brasil
eleitoradobrasil.nlargest(5, 'RELACAO_FE_MASC').T
```

```python
#Apenas eleitores que estão no Brasil
eleitoradobrasil.nsmallest(5, 'RELACAO_FE_MASC').T
```

```python
#Grafico da relação Feminino e Masculino
%matplotlib inline
eleitoradobrasil['RELACAO_FE_MASC'].plot.hist(bins=100)
```

```python
#Grafico de distribuiçao da relação feminino/ masculino
sns.distplot(eleitoradobrasil['RELACAO_FE_MASC'], bins=100, color='red', kde=False)
plt.title('Relação Eleitoras/ Eleitores', fontsize=18)
plt.xlabel('Eleitoras/ Eleitores', fontsize=14)
plt.ylabel('Frequencia', fontsize=14)
plt.axvline(1.0, color='black', linestyle='--')
```

```python
#Verificar com Grafico de distribuiçao da relação feminino/ masculino (municipios)
sns.swarmplot(data=eleitoradobrasil, x='NM_REGIAO', y='RELACAO_FE_MASC')
plt.axhline(1.0, color='black', linestyle='--')
```
